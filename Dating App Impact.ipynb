```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt   
import seaborn as sns     #import libraries

df = pd.read_csv('dating_app_behavior_dataset.csv')   #import dataset
df.head()  #Dataset Check
```

```python
x = df["bio_length"].sample(n=100, random_state=42)
y = df["likes_received"].sample(n=100, random_state=42)

# 3. Create the scatter chart
plt.scatter(x, y, alpha=0.3, s=10)  # alpha=0.3 makes points more transparent
plt.title("Bio Length vs. Likes Received")
plt.xlabel("Bio Length")
plt.ylabel("Likes Received")

# 4. Compute a best-fit line (linear regression)
# np.polyfit(x, y, 1) returns slope (m) and intercept (b) for a line y = m*x + b
m, b = np.polyfit(x, y, 1)

# 5. Plot the regression line
# we can plot the line using the same x-values and the formula y = m*x + b
plt.plot(x, m*x + b, color='red', linewidth=2)

# 6. Show the plot
plt.show()
```
